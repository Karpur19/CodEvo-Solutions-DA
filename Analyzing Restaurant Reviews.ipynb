```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load data
reviews_df = pd.read_csv("C:\\Users\\karpu\\OneDrive\\Desktop\\Yelp Restaurant Reviews.csv")

# Perform sentiment analysis on 'Review Text'
reviews_df['polarity'] = reviews_df['Review Text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Plot sentiment distribution
plt.figure(figsize=(10,5))
plt.hist(reviews_df['polarity'], bins=20, color='skyblue')
plt.title('Sentiment Distribution of Reviews')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Number of Reviews')
plt.show()
```

```python

# Plot rating distribution
plt.figure(figsize=(10,5))
reviews_df['Rating'].value_counts().sort_index().plot(kind='bar', color='coral')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()
```

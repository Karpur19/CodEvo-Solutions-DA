```python
Task 5: Analyzing Indian Government Open data

For this project, I have selected a dataset related to waste management in Vellore City, which is available in a CSV format. The dataset contains information about the number of residential households, non-residential establishments, and the coverage of door-to-door collection and source segregation in each ward of Vellore City.
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("C:\\Users\\karpu\\Downloads\\D70. Solid Waste Management - Household Coverage_2_0.csv")
```

```python
# Check for missing values
print(data.isnull().sum())

# Since there are no missing values, we can proceed with data cleaning

# Check for missing values
print(data.isnull().sum())

# Since there are no missing values, we can proceed with data cleaning

# Check for data types
print(data.dtypes)

# Convert 'Ward No.' to integer
data['Ward No.'] = data['Ward No.'].astype(int)

# Calculate percentage of households covered under door-to-door collection
data['Households_Door_to_Door_Collection_Percentage'] = (data['Total No. of Residential Households covered through door to door collection'] / data['Total No. of Residential Households']) * 100

# Calculate percentage of establishments covered under door-to-door collection
data['Establishments_Door_to_Door_Collection_Percentage'] = (data['Total No. of Non -residential establishments covered through door to door collection'] / data['Total No. of Non -residential establishments']) * 100

# Calculate percentage of households covered under source segregation
data['Households_Source_Segregation_Percentage'] = (data[' Households covered with source segregation'] / data['Total No. of Residential Households']) * 100

# Calculate percentage of establishments covered under source segregation
data['Establishments_Source_Segregation_Percentage'] = (data['Total No. of Non -residential establishments covered with source segregation'] / data['Total No. of Non -residential establishments']) * 100
```

```python
# Calculate average percentage of households covered under door-to-door collection
avg_households_door_to_door_collection_percentage = data['Households_Door_to_Door_Collection_Percentage'].mean()

# Calculate average percentage of households covered under source segregation
avg_households_source_segregation_percentage = data['Households_Source_Segregation_Percentage'].mean()

print(f'Average Households Door-to-Door Collection Percentage: {avg_households_door_to_door_collection_percentage}%')
print(f'Average Households Source Segregation Percentage: {avg_households_source_segregation_percentage}%')
```

```python
# Create a bar chart to show the average percentage of households covered under door-to-door collection and source segregation for each ward
plt.figure(figsize=(10,6))
sns.barplot(x='Ward No.', y='Households_Door_to_Door_Collection_Percentage', data=data)
plt.title('Average Percentage of Households Covered under Door-to-Door Collection by Ward')
plt.xlabel('Ward No.')
plt.ylabel('Average Percentage')
plt.show()
```

```python
# Create a line chart to show the trend of the average percentage of households covered under door-to-door collection and source segregation over time
plt.figure(figsize=(10,6))
sns.lineplot(x='Ward No.', y='Households_Door_to_Door_Collection_Percentage', data=data)
plt.title('Trend of Average Percentage of Households Covered under Door-to-Door Collection over Time')
plt.xlabel('Ward No.')
plt.ylabel('Average Percentage')
plt.show()
```

```python
# Create a scatter plot to show the relationship between the percentage of households covered under door-to-door collection and source segregation
plt.figure(figsize=(10,6))
sns.scatterplot(x='Households_Door_to_Door_Collection_Percentage', y='Households_Source_Segregation_Percentage', data=data)
plt.title('Relationship between Door-to-Door Collection and Source Segregation')
plt.xlabel('Door-to-Door Collection Percentage')
plt.ylabel('Source Segregation Percentage')
plt.show()
```

```python
# Create a categorical variable based on the Total No. of Non -residential establishments column
data['Establishment_Type'] = pd.cut(data['Total No. of Non -residential establishments'], bins=[0, 50, 100, 200, np.inf], labels=['Low', 'Medium', 'High', 'Very High'])

# Create a bar chart to show the average percentage of households covered under door-to-door collection and source segregation for each ward, grouped by the type of establishment
plt.figure(figsize=(10,6))
sns.barplot(x='Ward No.', y='Households_Door_to_Door_Collection_Percentage', hue='Establishment_Type', data=data)
plt.title('Average Percentage of Households Covered under Door-to-Door Collection by Ward and Establishment Type')
plt.xlabel('Ward No.')
plt.ylabel('Average Percentage')
plt.show()
```

```python
# Create a heatmap to show the correlation between the different variables
corr = data[['Households_Door_to_Door_Collection_Percentage', 'Households_Source_Segregation_Percentage', 'Establishments_Door_to_Door_Collection_Percentage', 'Establishments_Source_Segregation_Percentage']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation between Variables')
plt.show()
```
